```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
url_dataset = "https://raw.githubusercontent.com/M-ILHAM-197/prosaindata/main/airline-passengers.csv"
series = pd.read_csv(url_dataset, header=0, index_col=0)
```

```python
series
```

```python
def split_sequence(sequence, n_steps):
  X, y = list(), list()
  for i in range(len(sequence)):
  # find the end of this pattern
    end_ix = i + n_steps
    if end_ix > len(sequence)-1:
      break
    # gather input and output parts of the pattern
    seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
    X.append(seq_x)
    y.append(seq_y)
  return array(X), array(y)
```

```python
from numpy import array
passengers = series["Passengers"].values

# series = passengers
print(passengers.shape)
# transform to a supervised learning problem
X, y = split_sequence(passengers, 2)
print(X.shape, y.shape)
dfX = pd.DataFrame(X, columns=["Xt-2", "Xt-1"])
dfy = pd.DataFrame(y, columns=["Xt"])
# show each sample

df = pd.concat((dfX, dfy), axis = 1)
df
```

```python
dataY = df['Xt'].values
dataY
```
